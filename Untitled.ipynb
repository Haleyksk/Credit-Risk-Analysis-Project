```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

file_path = '/Users/hale/anaconda_projects/1e00ce7f-941f-47a6-a6ea-e6660da86289/credit_risk_dataset.csv'
df = pd.read_csv(file_path)

print("veri başarıyla yüklendi")
df.head()
```

```python
df.describe()
```

```python
df = df[df['person_age'] < 100 ]
df = df[df['person_emp_length'] < 60]
print("Yeni Max Yaş:", df['person_age'].max())
print("Yeni Max Çalışma yılı:", df['person_emp_length'].max())
```

```python
print(df.isnull().sum())
```

```python
df['loan_int_rate'] = df['person_emp_length'].fillna(0)
print("\nEksik veri temizliği tamamlandı!")
```

```python
print(df.isnull().sum())
```

```python
plt.figure(figsize=(10,6))
```

```python
# As a person's income increases, does the amount of credit they can take out also increase?
# x-axis: Income, y-axis: Loan Amount
sns.scatterplot(x='person_income', y='loan_amnt' , hue='loan_status' , data=df, alpha=0.5)
plt.title('The Relationship Between Income and Credit Amount')
plt.xlabel('Annual Income')
plt.ylabel('Loan Amount')
plt.show()
```

```python
# Does the risk increase as the credit rating (grade) worsens?
# x: loan grade , y: loan Status (0 or 1)
plt.figure(figsize=(10,6))
sns.barplot(x='loan_grade',y='loan_status',data=df, order=['A', 'B', 'C', 'D', 'E', 'F', 'G'],palette='viridis')
plt.title('Debt Default Rate Based on Credit Score (Risk Analysis)')
plt.xlabel(' loan grade (A: Best - G: Riskiest)')
plt.ylabel('Average Risk Rate (Default)')

plt.show()
```

```python
#Which data points follow each other?
numeric_df = df.select_dtypes(include=[np.number])
corr_matrix = numeric_df.corr()
plt.figure(figsize=(12,8))
sns.heatmap(corr_matrix, annot=True,cmap='coolwarm', fmt=".2f")
plt.title('Correlation Map Between Variables')
plt.show()
           
```

```python
from sklearn.preprocessing import LabelEncoder
categorical_cols = df.select_dtypes(include=['object']).columns
le = LabelEncoder()
for col in categorical_cols:
    df[col] = le.fit_transform(df[col])
    df.head()
```

```python
from sklearn.model_selection import train_test_split
x = df.drop('loan_status', axis=1)
y=df['loan_status']
x_train, x_test, y_train,y_test = train_test_split(x,y, test_size=0.2, random_state=42)
print(f"for education {x_train.shape[0]} line, for quise {x_test.shape[0]} line ready")
```

```python
from sklearn.ensemble import RandomForestClassifier
model = RandomForestClassifier(n_estimators=100, random_state=42)
model.fit(x_train,y_train)

y_pred = model.predict(x_test)
from sklearn.metrics import accuracy_score
success = accuracy_score(y_test,y_pred)
print(f"Model's Exam Score (Success Rate): %{success*100:.2f}")
```
